## Script para download de dados hist√≥ricos de cripyto moedas.

```python
from datetime import datetime, timedelta
import os
import requests
import time
```

```python
def write_data_url( response, file_name, path_dir ):
    with open( f"{path_dir}/{file_name}", "wb" ) as file:
        file.write( response.content )
    return
```

```python
def get_data_url( pair, interval, time_range ):
    now = datetime.now()
    date = now
    
    while now.year - date.year < time_range :
        year = str( date.year )
        month = date.month
        
        if month >= 10 and month <= 12:
            month = str( date.month )
        else:
            month = "0" + str( date.month )
            
        file_name = f"{pair}-{interval}-{year}-{month}.zip"
        path_dir = f"{pair}-{interval}"
        
        if not os.path.exists( path_dir ):
            os.mkdir( path_dir )
            
        data_url = f"https://data.binance.vision/data/spot/monthly/klines/{pair}/{interval}/{file_name}"
        try:
            response = requests.get( data_url )
            if response.status_code == 200:
                print( f"Download do arquivo {file_name} executado com sucesso!" )
        except:
            if response.status_code != 200:
                while response.status_code != 200:
                    print( f"Erro ao baixar arquivo {file_name}")
                    print( f"Tentando reconectar em 10 segundos" )
                    time.sleep( 10 )
                    response = requests.get( data_url )
                    if response.status_code == 200:
                        break
                    
            
        write_data_url( response, file_name, path_dir )
        date = date - timedelta( days = 30 )
    
    return
```

```python
usdt_pairs = [ 'BTCUSDT',  'ETHUSDT',  'XMRUSDT',  'XRPUSDT',   'BUSDUSDT',
               'DOGEUSDT', 'LTCUSDT',  'ETCUSDT',  'ZECUSDT',   'MANAUSDT' ]

api_intervals = [ "1m", "5m", "2h", "1d" ]

for pair in usdt_pairs:
    for interval in api_intervals:
        
        if interval == "1m":
            time_range = 1
            
        if interval == "5m":
            time_range = 2
        
        if interval == "2h":
            time_range = 3
            
        if interval == "1d":
            time_range = 5
                     
        get_data_url( pair, interval, time_range )
```
